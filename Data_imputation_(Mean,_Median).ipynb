Median imputation of missing values (for numerical columns **RCRI score** and **PreopEGFRMDRD**)

```python
# Read original dataset first
df = pd.read_csv("CARES_data_cleaned.csv")
df
```

```python
# calculate the median of RCRI score column, and replace missing values
median_RCRI = df['RCRI score'].median()
df['RCRI score'].fillna(median_RCRI, inplace = True)
df['RCRI score']
```

```python
# calculate the median of PreopEGFRMDRD column, and replace missing values
median_PreopEGFRMDRD = df['PreopEGFRMDRD'].median()
df['PreopEGFRMDRD'].fillna(median_PreopEGFRMDRD, inplace = True)
df['PreopEGFRMDRD']
```

```python
print("Mean RCRI score = " + str(mean_RCRI))
print("Median RCRI score = " + str(median_RCRI))
print("Mean PreopEGFRMDRD = " + str(mean_PreopEGFRMDRD))
print("Median PreopEGFRMDRD = " + str(median_PreopEGFRMDRD))
```

```python
df.describe()
```

The mean of the **RCRI score** decreased (from 0.322 to 0.225) after replacing missing values in the **RCRI score** column with the median.
