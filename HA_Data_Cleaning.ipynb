```python
url='https://github.com/ISSS623-AHA/ISSS623_2024/blob/main/Group_Project/Group_Project-SGH/CARES_data.xlsx?raw=true'

cares_df = pd.read_excel(url)
```

```python
# Remove empty rows
cares_df = cares_df.dropna(how='all')

# Remove the last row
cares_df = cares_df.iloc[:-1]

cares_df.tail()
```

```python
cares_df.describe()
```

```python
# Replace NA, #NULL! and BLANK with NaN

cares_df.replace('NA', np.nan, inplace=True)
cares_df.replace('#NULL!', np.nan, inplace=True)
cares_df.replace('BLANK', np.nan, inplace=True)
```

```python
cares_df.tail()
```

```python
cares_df[cares_df.duplicated(subset=None)]

# No duplicates found
```

```python
# Rename columns 'AGE' to 'Age' and 'GENDER' to 'Gender'
cares_df = cares_df.rename(columns={'AGE': 'Age', 'GENDER': 'Gender'})

# Capitalize only first character of each value in the 'Gender' column
cares_df['Gender'] = cares_df['Gender'].str.capitalize()

# Capitalize the first character of each value in the 'Anemia category' column
cares_df['Anemia category'] = cares_df['Anemia category'].str.capitalize()

# Convert GradeofKidneydisease g1 values to upper case
cares_df['GradeofKidneydisease'] = cares_df['GradeofKidneydisease'].apply(
    lambda x: x.upper() if isinstance(x, str) and x.lower() == 'g1' else x)
```

```python
# profile = ProfileReport(cares_df)

# profile.to_file(output_file='output_corrected.html')
```

```python
# Convert the 'Intraop' column to float type

cares_df['Intraop'] = cares_df['Intraop'].astype(float)
```

```python
# Check the datatype of each column
print(cares_df.dtypes)

# All the numeric columns are already in float type as compared to the data profiler
```

```python
boolean_cols = ['@30daymortality', 'Mortality', 'thirtydaymortality', 'CVARCRICategory', 'IHDRCRICategory',
                'CHFRCRICategory', 'DMinsulinRCRICategory', 'CreatinineRCRICategory', 'ICUAdmgt24h']
```

```python
# Label encoding for boolean values
def encode_boolean(value):
    if isinstance(value, str):
        value = value.strip().lower()
        if value == 'yes':
            return True
        elif value == 'no' or value == 'no death':
            return False
    return value

for col in boolean_cols:
    cares_df[col] = cares_df[col].apply(encode_boolean)
```

```python
print(cares_df.dtypes)

# Some boolean columns with NaN (empty) values will still appear as object datatype
```

```python
cares_df.head()
```

```python
categorical_cols = ['Gender', 'Anemia category', 'GradeofKidneydisease', 'AnaestypeCategory',
                    'PriorityCategory', 'TransfusionIntraandpostopCategory', 'AGEcategory',
                    'AGEcategoryOriginal', 'SurgRiskCategory', 'RaceCategory','GradeofKidneyCategory',
                    'Anemiacategorybinned', 'RDW15.7', 'ASAcategorybinned']
```

```python
# One-hot encoding for Categorical variables
cdf_one_hot=cares_df.copy()

for col in categorical_cols:
    if cares_df[col].isnull().any():  # Check if there are any missing values in the column
        one_hot = pd.get_dummies(cares_df[col], dummy_na=True)
    else:
        one_hot = pd.get_dummies(cares_df[col], dummy_na=False)  # No 'NA' column if no missing values

    # retain a column to indicate NaN only if it exists
    one_hot.columns = one_hot.columns.fillna('NA')
    one_hot.columns = [col + '_' + a for a in one_hot.columns]
    cdf_one_hot = cdf_one_hot.join(one_hot)
```
